# Visualizing Options Payoff

My main objective in this Notebook is to show the Opstrat library and the powerfull resources provided by the library to visualize options strategies payoffs

# 1.0 Import Library

with ! pip install opstrat we can download the library, the requirements are pandas, numpy, matplotlib, seaborn, yfinance already installed

```python
import opstrat as op
```

# 2.0 Default Functions

The method single plotter plot to us an Call with 102(Monetary Units) Strike Price in a stock which have a spot price of 100 (Monetary Units), this option costs 10 (Monetary Units). The graph to us a payoff range of +- 5% and the position of the contract is Long!

```python
op.single_plotter()
```

.

.

# 3.0 Input Parameters

It´s important to know how to set our own contract parameters to analyse it. For example I will set an option with 260 (Monetary Units) as strike , Spot Price of 267 (Monetary Units) , a Put Option, which have an short position and a premium of 3.4 (Monetary Units)

```python
op.single_plotter(spot=267, strike=260, op_type='p', tr_type='s', op_pr=3.4)
```

It´s import to know that the breakeven in that condition is calculated by (Strike Price - Premium Collected) so in that case 256.60 is the Breakeven!
